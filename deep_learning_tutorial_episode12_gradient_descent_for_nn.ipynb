```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_pickle("G:\\My Drive\\Projects\\110_JetscapeMl\\hm.jetscapeml.data\\simulation_results\\jet_ml_benchmark_config_01_to_09_alpha_0.2_0.3_0.4_q0_1.5_2.0_2.5_MMAT_MLBT_size_1000_shuffled.pkl")

type(df)
dataset_x=df[0]
dataset_y=df[1]
dataset_x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(dataset_x,dataset_y, test_size=0.2,random_state=25)
```

```python
x_train.shape,len(x_train)
```

```python
x_max=max(x_train.max(),x_test.max())
x_max
```

```python
x_train_flatten=x_train.reshape(len(x_train),32*32)
x_test_flatten=x_test.reshape(len(x_test),32*32)
x_train_scaled=x_train_flatten.copy()
x_test_scaled=x_test_flatten.copy()
x_train_scaled=x_train_scaled/x_max
x_test_scaled=x_test_scaled/x_max
```

```python
from sklearn.preprocessing import OneHotEncoder


y_train_0_categorical = np.array(y_train[:,0]).reshape(-1, 1)
y_test_0_categorical = np.array(y_test[:,0]).reshape(-1, 1)

encoder = OneHotEncoder(sparse_output=False)
y_train_0_categorical_encoded = encoder.fit_transform(y_train_0_categorical)
y_test_0_categorical_encoded = encoder.transform(y_test_0_categorical)
y_test_0_categorical_encoded
```

```python
model=keras.Sequential([
    keras.layers.Dense(2,input_shape=(32*32,),activation='sigmoid',kernel_initializer='ones',bias_initializer='zeros')
])
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics='accuracy')
model.fit(x_train_scaled,y_train_0_categorical_encoded,epochs=5000)
```

```python
model.evaluate(x_test_scaled,y_test_0_categorical_encoded)
```

```python
def log_loss(y_true,y_predicted):
    epsilon=1e-15
    y_predicted_new=[max(i,epsilon) for i in y_predicted]
    y_predicted_new=[min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new=np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
def sigmoid_numpy(X):
    return 1/(1+np.exp(-X))
sigmoid_numpy(np.array([12,0,1]))
```

```python
def gradient_descent(event_image_flatten,y_true,epoch,threshold):
    #wi ,bias
    w=np.ones(len(event_image_flatten[0]))
    # bias=np.zeros(len(event_image_flatten[0]))
    bias=0
    rate=0.5
    n=len(event_image_flatten)
    for i in range (epoch):
        # print(event_image_flatten.shape)
        weighted_sum=np.array([])
        for event_image_flatten_item in  event_image_flatten:
            current_weighted_sum=np.dot(event_image_flatten_item,w)+bias
            # print(current_weighted_sum)
            # break
            weighted_sum=np.append(weighted_sum,np.array([current_weighted_sum]))
            # print(current_weighted_sum)
            # print(current_weighted_sum.shape)
            # break
        # weighted_sum=np.dot(event_image_flatten,w)+bias
        # print(weighted_sum)
        # print(weighted_sum.shape)
        # break
        y_predicted=sigmoid_numpy(weighted_sum)
        
        # print(y_predicted.shape)
        # print(y_true.shape)
        # break
        loss=log_loss(y_true,y_predicted)
        # print(loss)
        # print(event_image_flatten.shape)
        # difference=y_predicted-y_true
        # print(difference.shape)
        
        w_d=(1/n)*np.dot(np.transpose(event_image_flatten),(y_predicted-y_true))
        # print(w_d.shape)
        b_d=np.mean(y_predicted-y_true)
        w=w-rate*w_d
        bias=bias-rate*b_d
        print(f'Epoch:{i},w:{w},bias:{bias},loss:{loss}')
        if loss<=threshold:
            break

    return w,bias
```

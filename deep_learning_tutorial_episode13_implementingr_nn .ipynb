```python
def log_loss(y_true,y_predicted):
    epsilon=1e-15
    y_predicted_new=[max(i,epsilon) for i in y_predicted]
    y_predicted_new=[min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new=np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
def sigmoid_numpy(X):
    return 1/(1+np.exp(-X))
sigmoid_numpy(np.array([12,0,1]))
```

```python
class myNN:
    def __init__(self):
        self.w=np.ones(1024)
        self.bias=0
        pass
    def fit(self,X,y,epoch,loss_threshold):
        self.w,self.bias=self.gradient_descent(X,y,epoch,loss_threshold)
    def predict(self,X_test):
        weighted_sum=np.dot(X_test,self.w)+self.bias
        return sigmoid_numpy(weighted_sum)

    def gradient_descent(self,event_image_flatten_items,y_true,epoch,loss_threshold):
        #wi ,bias
        w=np.ones(1024)
        bias=0
        rate=0.5
        n=len(event_image_flatten_items)
        for i in range (epoch):
            weighted_sum=np.dot(w,event_image_flatten_items.T)+bias
            # print(weighted_sum.shape)
            y_predicted=sigmoid_numpy(weighted_sum)
            # print(y_predicted.shape)
            loss=log_loss(y_true,y_predicted)
            # print(loss,loss.shape)

            # print((y_predicted-y_true).shape)
            # print(np.transpose(event_image_flatten_items).shape)
            w_d=(1/n)*np.dot(np.transpose(event_image_flatten_items),(y_predicted-y_true))
            bias_d=np.mean(y_predicted-y_true)
            w=w-rate*w_d
            bias=bias-rate*bias_d
            if i%50==0:
                print(f'Epoch:{i},w:{w},bias:{bias},loss:{loss}')
            if loss<=loss_threshold:
                print(f'Epoch:{i},w:{w},bias:{bias},loss:{loss}')
                break

        return w,bias
    
```
